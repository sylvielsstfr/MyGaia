```python
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
import healpy as hp 
import numpy as np
```

```python
file1="MOC-I_345_gaia2-512.fits"
file1="densityMap-I_345_gaia2-64.hpx"
file2="densityMap-I_345_gaia2-128.hpx"

file3="densityMap-I_345_gaia2-256.hpx"
file5="densityMap-I_345_gaia2-1024.hpx"
file4="densityMap-I_345_gaia2-512.hpx"
```

# densityMap-I_345_gaia2-64.hpx

```python
nside1=64
npix1=hp.nside2npix(nside1)
map1 = hp.read_map(file1)
resol1= hp.pixelfunc.nside2resol(nside1, arcmin=True)
```

```python
hp.mollview(map1,norm='log',title='Star density',cmap="jet",xsize=2000)
hp.graticule(color="w")
```

# densityMap-I_345_gaia2-128.hpx

```python
nside2=128
npix2=hp.nside2npix(nside2)
map2 = hp.read_map(file2)
resol2= hp.pixelfunc.nside2resol(nside2, arcmin=True)
```

```python
imap2=map2.astype(np.int64)
```

```python
hp.mollview(imap2,norm='log',title='Star density',cmap="jet",xsize=2000)
hp.graticule(color="w")
```
